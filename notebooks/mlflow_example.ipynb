```python
import mlflow
import os
import mlflow
from dotenv import load_dotenv

load_dotenv()
tracking_uri = os.getenv("MLFLOW_TRACKING_URI")
if not tracking_uri:
    raise ValueError("Không tìm thấy biến môi trường 'MLFLOW_TRACKING_URI'. Vui lòng kiểm tra lại cấu hình.")
print(f"-> Đang kết nối tới MLflow tại: {tracking_uri}")

experiment_name = "DataMiningModels"
try:
    mlflow.set_experiment(experiment_name)
    experiment = mlflow.get_experiment_by_name(experiment_name)
    print(f"-> Đã trỏ tới Experiment: '{experiment_name}' (ID: {experiment.experiment_id})")
except Exception as e:
    print(f"Lỗi khi set experiment: {e}")
```
